# Replicating Teacher's Graph with Matplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Replicatiing the relation between wealth and probability, therefore the probability distribution

wk1 = pd.read_csv('wktime.T001.his', delimiter=' ', header=None) #Importing the file as a data frame
wk2 = pd.read_csv('wktime.T004.his', delimiter=' ', header=None) #Import the 2nd file as a data frame


plt.plot(wk1[0], wk1[1], color='r', label='Time 1') #Drawing the first plot graph

plt.plot(wk2[0], wk2[1], color='y', label='Time 2') #Drawing the second plot graph

#Writting the titles of the axes
plt.xlabel("Wealth")
plt.ylabel("Probability")
plt.title("Probability Distribution")
plt.xscale("log")                      # axis X scale
plt.yscale("log")                      #Axis Y scale
plt.xlim((0.000001, 100.0))            # axis X limit
plt.ylim((0.000001, 100000.0))  
plt.tight_layout()






plt.show()
```

# Free graph

```python
#Random graph to make a linear regression analysis 

#Generating random dots with numpy
np.random.seed(0)
x = np.random.rand(150, 1)  # 150 dots generated

y = 1 + 5*x + np.random.rand(150, 1)   #The form the dots will follow

#Now generate the sizes of the graph
plt.figure(figsize=(10, 5), dpi=80) 
plt.ylim(1, 7)   #Limits y
plt.xlim(0, 1)   # limits x
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Introduction to linear regression with MatPlotLib")
plt.scatter(x, y, s=15, color="g", label="Data")  #The form and color of the dots
plt.legend()   # The legend
plt.show()
```
